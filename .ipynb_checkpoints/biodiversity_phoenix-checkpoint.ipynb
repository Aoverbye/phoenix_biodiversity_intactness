```python
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
import rioxarray as rioxr

from pystac_client import Client  # To access STAC catalogs

import planetary_computer  # To sign items from the MPC STAC catalog 

from IPython.display import Image  # To nicely display images

# Access MPC catalog
catalog = Client.open(
    "https://planetarycomputer.microsoft.com/api/stac/v1",
    modifier=planetary_computer.sign_inplace,
)

biodiversity = catalog.get_child('io-biodiversity')
biodiversity
```

```python
# Temporal range of interest
time_range = "2020-01-01/2020-12-31"

# Pheonix bounding box 
bbox = {
    "type": "Polygon",
    "coordinates": [
        [
            [-112.826843, 32.974108],  # Bottom-left corner
            [-112.826843, 33.863574],  # Top-left corner
            [-111.184387, 33.863574],  # Top-right corner
            [-111.184387, 32.974108],  # Bottom-right corner
            [-112.826843, 32.974108],  # Closing the polygon
        ]
    ],
}
    
# Catalog search
search = catalog.search(
    collections = ['io-biodiversity'],
    intersects = bbox,
    datetime = time_range)
search

# Retrieve search items
items = search.item_collection()
len(items)
```

```python


# Search for items in the 'io-biodiversity' collection
search = catalog.search(collections=["io-biodiversity"], bbox=bbox, datetime=time_range)
items = search.item_collection()

# Convert the search results to a GeoDataFrame
df = gpd.GeoDataFrame.from_features(items.to_dict(), crs="epsg:4326")

# Inspect the GeoDataFrame
print(df.head())



# Access and read the COG data
for item in items:
    # Sign the asset URL using Planetary Computer's helper function
    signed_href = planetary_computer.sign(item.assets["data"].href)
    
    # Optionally print the URL for confirmation
    print(signed_href)
    
    # Use rasterio or another library to read and analyze the data
    # Example: Read the raster using rasterio
    import rasterio
    with rasterio.open(signed_href) as src:
        data = src.read(1)  # Read the first band
        print(data.shape)   # Inspect the shape of the raster
```

```python
# Import necessary libraries
import rasterio
import rasterio.plot
import matplotlib.pyplot as plt
from rasterio.plot import show
import geopandas as gpd
from shapely.geometry import box

# Define the Phoenix subdivision bounding box as a GeoDataFrame
bbox = [-112.826843, 32.974108, -111.184387, 33.863574]
phoenix_bbox = gpd.GeoDataFrame(
    {"geometry": [box(*bbox)]}, crs="EPSG:4326"
)

# Retrieve the first BII item from the search results
item = items[0]
signed_href = planetary_computer.sign(item.assets["data"].href)

# Read the raster data
with rasterio.open(signed_href) as src:
    data = src.read(1)  # Read the first band
    transform = src.transform

# Create a plot
fig, ax = plt.subplots(figsize=(10, 10))

# Plot the raster data
rasterio.plot.show(
    data, transform=transform, ax=ax, cmap="viridis", title="Biodiversity Intactness Index (BII)"
)

# Overlay the Phoenix subdivision bounding box
phoenix_bbox.boundary.plot(ax=ax, edgecolor="red", linewidth=2, label="Phoenix Subdivision")

# Add a legend and grid
ax.legend(loc="upper right")
ax.grid(True)

# Show the map
plt.show()
```
